```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
ingredient = "barley"
url = "https://feedtables.com/content/" + ingredient
request = requests.get(url)
request
```

```python
request.content
```

```python
tables = pd.read_html(request.content)
tables
```

```python
table_titles = BeautifulSoup(request.content).find_all("caption")
table_titles
```

```python
table_titles_list = [table_titles[n].text.strip() for n in range(len(table_titles))]
```

```python
table_titles_list
```

```python
main_content = pd.concat ([tables[n] for n in range(len(tables))],
                          keys=table_titles_list)
main_content
```

```python
import os
newpath = r'C:/' + ingredient 
if not os.path.exists(newpath):
    os.makedirs(newpath)
main_content.to_excel("C:/" + ingredient + "/" + ingredient + ".xlsx")
# Now your excel file has been saved in the folder of it's name in drive C
```
